```python
print("Welcome to the data analyzer and transformer program.")

lst = []

def enter_data():
    choice = int(input("Enter 1 for 1D list or 2 for 2D list: "))

    if choice == 1:
        for i in range(int(input("Enter number of elements: "))):
            lst.append(int(input(f"Enter element {i+1}: ")))

    else:
        r = int(input("Enter rows: "))
        c = int(input("Enter columns: "))
        for i in range(r):
            row = []
            for j in range(c):
                row.append(int(input(f"Enter element {i+1},{j+1}: ")))
            lst.append(row)

    print("Your data:", lst)


def data_summary():
    data = [j for i in lst for j in i] if isinstance(lst[0], list) else lst

    print("Total elements:", len(data))
    print("Sum:", sum(data))
    print("Min:", min(data))
    print("Max:", max(data))
    print("Average:", sum(data)/len(data))


def factorial(n):
    f = 1
    for i in range(2, n+1):
        f *= i
    return f


def filter_data():
    data = [j for i in lst for j in i] if isinstance(lst[0], list) else lst
    t = int(input("Enter threshold value: "))
    print("Filtered Data:", [x for x in data if x > t])


def sort_data():
    if isinstance(lst[0], list):
        for i in lst:
            i.sort()
            print(i)
    else:
        ch = int(input("1. Ascending  2. Descending: "))
        print(sorted(lst, reverse=(ch == 2)))


def return_dataset():
    data = [j for i in lst for j in i] if isinstance(lst[0], list) else lst
    return min(data), max(data), sum(data)/len(data)


while True:
    print("\n1.Enter data\n2.Summary\n3.Factorial\n4.Filter\n5.Sort\n6.Stats\n7.Exit")
    n = int(input("Enter choice: "))

    if n == 1:
        enter_data()
    elif n == 2:
        data_summary()
    elif n == 3:
        num = int(input("Enter number: "))
        print("Factorial:", factorial(num))
    elif n == 4:
        filter_data()
    elif n == 5:
        sort_data()
    elif n == 6:
        mn, mx, avg = return_dataset()
        print("Min:", mn, "Max:", mx, "Average:", avg)
    elif n == 7:
        print("Thank you!")
        break
        
```
